```python
#Group By
```

```python
import numpy as np
import pandas as pd
```

```python
data = {'Company' :['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person' :['A','B','C','D','E','F'],
       'Sales' :[200,300,120,500,348,230]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
byComp = df.groupby('Company')
```

```python
byComp
```

```python
byComp.mean()
```

```python
byComp.sum()
```

```python
byComp.std()
```

```python
byComp.sum().loc['FB']
```

```python
df.groupby('Company').sum().loc['FB']
```

```python
df.groupby('Company').count()
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').min()
```

```python
df.groupby('Company').describe()
```

```python
df.groupby('Company').describe().transpose()
```

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'A' : ['A0','A1','A2','A3'],
                    'B' : ['B0','B1','B2','B3'],
                    'C' : ['C0','C1','C2','C3'],
                    'D' : ['D0','D1','D2','D3']},
                    index = [0,1,2,3])
```

```python
df2 = pd.DataFrame({'A' : ['A4','A5','A6','A7'],
                    'B' : ['B4','B5','B6','B7'],
                    'C' : ['C4','C5','C6','C7'],
                    'D' : ['D4','D5','D6','D7']},
                    index = [4,5,6,7])
```

```python
df3 = pd.DataFrame({'A' : ['A8','A9','A10','A11'],
                    'B' : ['B8','B9','B10','B11'],
                    'C' : ['C8','C9','C10','C11'],
                    'D' : ['D8','D9','D10','D11']},
                    index = [8,9,10,11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
#concatenation
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3], axis = 1)
```

```python
left = pd.DataFrame({'key' : ['k0','k1','k2','k3'],
                    'A' : ['A0','A1','A2','A3'],
                    'B' : ['B0','B1','B2','B3']})


right = pd.DataFrame({'key' : ['k0','k1','k2','k3'],
                    'C' : ['C0','C1','C2','C3'],
                    'D' : ['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
# Merging
```

```python
pd.merge(left,right,how = 'inner',on='key')
```

```python
pd.merge(left,right, how = 'outer', on = 'key')
```

```python
# Joining
```

```python
left = pd.DataFrame({'A' : ['A0','A1','A2'],
                    'B' : ['B0','B1','B2']},
                   index = ['K0','K1','K2'])


right = pd.DataFrame({'C' : ['C0','C2','C3'],
                    'D' : ['D0','D2','D3']},
                    index = ['K0','K2','K3'])
```

```python
left.join(right)
```
